```python
#import lib 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
Cities_cali=pd.read_csv("california_cities.csv")
Cities_cali.head()
```

```python
latd,logd=Cities_cali["latd"],Cities_cali["longd"]
```

```python
fig,ax=plt.subplots(figsize=(15,10))

ax.set_title("California Cities: Population and Area Distribution",color="green",fontsize=40)
ax.set_ylabel("Longitude",fontsize=25,color="blue")
ax.set_xlabel("Latitude",fontsize=25,color="blue")
area,population=Cities_cali["area_total_km2"],Cities_cali["population_total"]
cali_cities=ax.scatter(latd,logd,alpha=0.4,s=area,c=np.log10(population),cmap="gist_rainbow_r")
ax.axis("tight")
fig.colorbar(cali_cities,label="log$_{10}$(Population)");
area_range=[50,150,250,500]
for area in area_range:
    ax.scatter([],[],label=str(area)+" km$^2$ ",s=area,c="k",alpha=0.4)
    
ax.legend();


```
